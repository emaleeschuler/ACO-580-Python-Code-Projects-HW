Question 1:

The half life of Cesium 90, a dangerous radioactive isotope released in the Fukushima reactor accident, is 30 years. That is, half of its radioactive material will decay every 30 years. Rearrange the following lines to produce a program that computes when 99 percent of the material is gone.

```python
year = 0
percentage = 100

while percentage > 1.0 :
    percentage = percentage * 0.5
    year = year + 30
print(year)
```

Question 2:

Write a while loop that prints all positive numbers that are divisible by 10 and less than a given number n. For example, if n is 100, print 10 20 30 40 50 60 70 80 90, one value per line.

```python
n = int(input("n: "))
i = 10
while i < n :
   print(i)
   i += 10
```

Question 3:

Trace through the following loop.

```python
n = 1796
count = 0;
while n > 0 :
   digit = n % 10
   if digit == 6 or digit == 7 :
      count = count + 1
   n = n // 10
print(count)
```

Question 4: 

Write a program that prints all digits of any integer in reverse order, one per line.

```python
n = int(input("Enter an integer: "))

# Convert the number to a string and reverse it
number_str = str(n)
reversed_number_str = number_str[::-1]

# Print each digit of the reversed number, one per line
for digit in reversed_number_str:
    print(digit)

   
## TODO: Print the digits of n in reverse order
```
