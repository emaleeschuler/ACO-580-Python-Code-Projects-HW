Question 2:

Complete the following program that multiplies each element of a list by five.

```python
values = [2,3,5,7,11]

# Multiply each element of the values list by 5.

def multiply(values, factor):
   for i in range(len(values)): 
      values[i] = values[i] * factor

multiply(values, 5)

# Print the resulting list:

for element in values:
    print(element)
```

```python
values = [17, 25, 5, 30, 100, 96, 28, 5, 14, 15, 30]

#Find the maximum value. 

max(values)

#Remove the maximum value from the list

values.remove(max(values))

#Print the resulting list
print(values)
```

Question 4:

Complete the following program that creates a duplicate version of a list but with the elements stored in reverse order from the original.

```python
origValues = [17, 25, 5, 30, 100, 96, 48, 5, 14, 30]
newValues = []

#Add the elements from the origValues list to the newValues list in reverse order.
for value in reversed(origValues) : 
   newValues.append(value)

#Print new values

print(newValues)
```
