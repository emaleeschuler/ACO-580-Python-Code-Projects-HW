Question 4:

Complete this program that prints the position of the last uppercase letter in a string. Print -1 if the word has no uppercase letters.

```python
word = input("word: ")
position = len(word) - 1  # Start from the end of the word
found = False  # Initialize found to False assuming no uppercase letters are found

# Look through each character in the word from right to left
while position >= 0:
    ch = word[position]
    if ch.isupper():
        found = True
        break  # Exit the loop if uppercase letter found
    position -= 1  # Move to the previous character

# Print the position of the last uppercase letter or -1 if none is found
if found:
    print("Position:", position)
else:
    print("Position:", -1)
```

Question 5:

Walk through the following code that counts how many adjacent duplicates are in a sequence of non-negative integer inputs, using a negative value as the sentinel. Assume the inputs are 1 2 2 4 4 4 3 -1.

```python
duplicates = 0
value = int(input("Enter a value: "))
while value >= 0 :
   previous = value
   value = int(input("Enter a value: "))
   if value == previous :
      duplicates = duplicates + 1

print(duplicates)
```
