Question 4:

Consider the follow program that prints a page number on the left or right side of a page. Define and use a new function, isEven, that returns a Boolean to make the condition in the if statement easier to understand.

```python
def main():
    page = int(input("Enter page number: "))
    
    def isEven(page_number):
        return page_number % 2 == 0

    if isEven(page):
        print(page)
    else:
        print("%60d" % page)

main()
```

Question 5:

Consider the following program that counts the number of spaces in a user-supplied string. Modify the program to define and use a function, countSpaces, instead.

```python
def main():
    userInput = input("Enter a string: ")
   
    def countSpaces(userInput):
        spaces = 0
        for char in userInput:
            if char == " ":
                spaces = spaces + 1
        return spaces
   
    result = countSpaces(userInput)
    print(result)

main()
```
